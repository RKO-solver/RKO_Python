```python
import kp as kp
import os
```

```python
if __name__ == "__main__":
    env = kp.KnapsackProblem(os.path.join(kp.current_directory,'kp50.txt'))
    logger = kp.FileLogger(os.path.join(kp.current_directory,'results.txt'), reset=True)
    solver = kp.RKO(env, logger=logger)
    solver.solve(time_total=30, brkga=1, lns=1, vns=1, ils=1, sa=1, pso=1, ga=1, runs=2)
```

```python
kp.HistoryPlotter.plot_convergence("results.txt").show()
```
